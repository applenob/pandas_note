## 数据集的合并（Merge）和连接（join）

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df1 = DataFrame({'key':['b','b','a','c','a','a','b'],'data1':range(7)})
df1
```

```python
df2 = DataFrame({'key':['a','b','d'],'data2':range(3)})
df2
```

Merge相当于数据库中的联接。包括内联、左联接、右联接、外联。
- 内联：select * from df1 inner join df2 on df1.key=df2.key
- 外联：select * from df1 full outer join df2 on df1.key=df2.key
- 左联接：select * from df1 left outer join df2 on df1.key=df2.key
- 右联接：select * from df1 right outer join df2 on df1.key=df2.key

```python
pd.merge(df1,df2) #没有指定，就会把重叠列的列名当做键
```

```python
pd.merge(df1,df2,on='key') #显式指定键名
```

```python
df3 = DataFrame({'keyl':['b','b','a','c','a','a','b'],'data1':range(7)})
df3
```

```python
df4 = DataFrame({'keyr':['a','b','d'],'data2':range(3)})
df4
```

```python
pd.merge(df3,df4,left_on='keyl',right_on='keyr') #键名不一样时，分别指定
```

```python
pd.merge(df1,df2,how='outer') #外联
```

```python
pd.merge(df1,df2,how='left') #左联接
```

```python
pd.merge(df1,df2,how='right') #右联接
```

## 轴向联接

```python
arr = np.arange(12).reshape((3,4))
arr
```

```python
np.concatenate([arr,arr],axis=1) #np的concatenate
```

```python
np.concatenate([arr,arr],axis=0)
```

```python
s1 = Series([0,1],index=['a','b'])
s2 = Series([2,3,4],index=['c','d','e'])
s3 = Series([5,6],index=['f','g'])
pd.concat([s1,s2,s3]) #pd的condat
```

```python
pd.concat([s1,s2,s3],axis=1) #默认结果是并集
```

```python
s4 = pd.concat([s1*5,s3])
s4
```

```python
pd.concat([s1,s4],axis=1)
```

```python
pd.concat([s1,s4],axis=1,join='inner') #指定为交集
```

```python
pd.concat([s1,s4],axis=1,join_axes=[['a','c','b','e']]) #指定索引
```

```python
result = pd.concat([s1,s1,s3],keys=['one','two','three'])
result #保留合并前的区域，并命名
```

```python
result.unstack()
```

## 合并部分数据

```python
a = Series([np.nan,2.5,np.nan,3.5,4.5,np.nan],
           index=['f','e','d','c','b','a'])
b = Series(np.arange(len(a),dtype=np.float64),
            index=['f','e','d','c','b','a'])
b[-1] = np.nan
```

```python
a
```

```python
b
```

```python
np.where(pd.isnull(a),b,a) #相当于if(pd.isnull(a)),then(b),else(a)
```

```python
a.combine_first(b) #实现和上句相同的功能，可以理解为参数对象中的数据为调用者对象打补丁
```

## 映射

```python
data = DataFrame({'food':['bacon','pulled pork','bacon',
                          'pastrami','corned beef','Bacon',
                          'pastrami','honey ham','nova lox'],
                  'ounces':[4,3,12,6,7.5,8,3,5,6]})
data
```

```python
meat_to_animal={
    'bacon':'pig',
    'pulled pork':'pig',
    'pastrami':'cow',
    'corned beef':'pig',
    'honey ham':'pig',
    'nova lox':'salmon'
}           #食物到动物的映射
```

```python
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```

```python
data['food'].map(lambda x: meat_to_animal[x.lower()])
```

## 离散化和面元（bin）划分

```python
ages = [20,22,25,27,21,23,37,31,61,45,41,32]
```

```python
bins = [18,25,35,60,100]
```

```python
cats = pd.cut(ages,bins)
cats
```

```python
cats.labels
```

```python
cats.codes
```

```python
cats.levels
```

```python
cats.categories
```

```python
group_names=['Youth','YoungAdult','MiddleAged','Senior']
pd.cut(ages,bins,labels=group_names)
```

```python
data = np.random.rand(20)
pd.cut(data,4,precision=2)
```

```python
data= np.random.randn(1000)#正态分布
cats= pd.qcut(data,4) #按4分位数进行切割，得到大小几乎相等的面元
```

```python
cats
```

```python
pd.value_counts(cats)
```

```python
mnames = ['movie_id','title','genres']
movies = pd.read_table('G:\projects\pycharm\pandas_demo\ch02\movielens\movies.dat',
                       sep='::',header=None,
                       names=mnames)
movies[:10]
```

```python
genre_iter = (set(x.split('|')) for x in  movies.genres)
genres = sorted(set.union(*genre_iter))
genres
```

```python
dummies = DataFrame(np.zeros((len(movies),len(genres))),columns=genres)
for i,gen in enumerate(movies.genres):
    dummies.ix[i,gen.split('|')]=1
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.ix[0]
```
