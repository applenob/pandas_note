## 逐块读取文本文件

```python
result = pd.read_csv('ch06/ex6.csv')
result
```

```python
pd.read_csv('ch06/ex6.csv',nrows=5)  #只想读取几行
```

```python
chunker = pd.read_csv('ch06/ex6.csv',chunksize=1000) #返回的对象可以用来逐块迭代
chunker
```

```python
tot = Series([])                       #逐块值计数，再相加
for piece in chunker:
    tot = tot.add(piece['key'].value_counts(),fill_value=0)
tot = tot.order(ascending=False)
tot[:10]
```
